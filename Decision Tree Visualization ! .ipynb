```python
from dtreeviz.trees import *
from dtreeviz.trees import explain_prediction_path , dtreeviz
import graphviz.backend as be   #IMP 
#from IPython.display import Image, display_svg, SVG
```

```python
from sklearn.datasets import load_iris, load_boston
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor
```

## Decision Tree For Classification



```python
# Loading data 
iris = load_iris()
X_train = iris.data
y_train = iris.target

# inintializing and loading the decision tree 
dtc = DecisionTreeClassifier(max_depth=3)
dtc.fit(X_train,y_train)
```

## Plotting Decision Tree using Sklearn 

```python
from sklearn.tree import plot_tree
plot_tree(dtc)
```

## 1. Plotting using dtreeviz

```python
viz = dtreeviz(
    dtc,
    X_train,
    y_train,
    feature_names= iris.feature_names,
    class_names= ['setosa', 'versicolor', 'virginica'], 
    scale = 1.5
)
viz
```

## 2. Decision Tree For Regression

```python
data = load_boston()
XR_train = data.data
yR_train = data.target
dtr = DecisionTreeRegressor(max_depth=2)
dtr.fit(XR_train, yR_train)
```

```python
viz = dtreeviz(
    dtr , XR_train, yR_train, target_name = 'price',  feature_names = data.feature_names , scale = 1
)
viz
```

## 3. Decision Tree horizontally 

```python
# orientation = 'LR'
viz = dtreeviz(
    dtr , XR_train, yR_train, target_name = 'price',  feature_names = data.feature_names , scale = 1 , orientation = 'LR'
)
viz
```

## 4. Showing decision path 

```python
# Loading data 
iris = load_iris()
X_train = iris.data
y_train = iris.target

# inintializing and loading the decision tree 
dtc = DecisionTreeClassifier(max_depth=3)
dtc.fit(X_train,y_train)

# fitting the X_test
X_test = X_train[135, :]
y_test = iris.target_names[y_train[135] ]

```

```python
viz = dtreeviz(
    dtc,
    X_train,
    y_train,
    feature_names= iris.feature_names,
    class_names= ['setosa', 'versicolor', 'virginica'], 
    scale = 1.5,
    X = X_test
)
print()
print()
print()
print( "the input data was = " ,str(X_test) , "the predicted output was " , str(y_test) )
viz
# here the traing data we are back tracking it to visualize it ! 
```

## 5. Show Node Labels 

```python
#show_node_labels = True 

viz = dtreeviz(
    dtc,
    X_train,
    y_train,
    feature_names= iris.feature_names,
    class_names= ['setosa', 'versicolor', 'virginica'], 
    scale = 1.5,
    X = X_test, 
    show_node_labels = True 
)
viz
```

## 6 Feature Importance 

```python
## Feature importance of the data to be predicted 

explain_prediction_path(tree_model=dtc, x= X_test, x_data= X_train, y_data= y_train ,
                        feature_names= iris.feature_names, explanation_type='sklearn_default') 
```
